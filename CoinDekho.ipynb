```python
from ultralytics import YOLOv10
model=YOLOv10("/content/runs/detect/train/weights/best.pt")
model(source="/content/indian-coin-1/valid/images",conf=0.25,save=True)
```

```python
import glob
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
import cv2

images=glob.glob("/content/runs/detect/predict/*.jpg")
numofimg=images[:10]
fig,axes=plt.subplots(2,5,figsize=(20,10))

for i,axs in enumerate(axes.flat):
  if i<len(numofimg):
    img=pltimg.imread(numofimg[i])
    axs.imshow(img)
    axs.axis("off")
  else:
    axs.axis("off")
plt.tight_layout()
plt.show()
```

```python
image_path="/content/mytest.png"
res=model.predict(source=image_path,imgsz=640,conf=0.15)
annimg=res[0].plot()
```

```python
plt.imshow(annimg[:,:,::-1])
```

```python
detects=res[0].boxes.data
prntclass=[model.names[int(cls)] for cls in detects[:,5]]
```

```python
counts={}
for i in prntclass:
  if i in counts:
    counts[i]+=1
  else:
    counts[i]=1
```

```python
print(counts)
```

```python
sum=0
for i in counts:
  sum+=counts[i]*int(i)
print(f"Total Sum of Coins : {sum}")
```
